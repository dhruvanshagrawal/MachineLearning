```python
def convert_sqft_to_number(x):
    tokens = x.split("-")
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None

df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_number)
```

```python

df['total_sqft'][30]
```

```python
df['bath'].count()
```

```python
df['price_per_sqft'] = df['price']*100000/df['total_sqft']
```

```python
df.head()
```

```python
df = df[df['price_per_sqft'] >= 400]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = df.drop(columns="location", axis=1)
```

```python
df.head()
```

```python
df= df.drop(columns=['society','availability','balcony','bath','area_type'], axis=1)
```

```python
df.head()
```

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
l_regressor = LinearRegression()
l_regressor.fit(X_train, y_train)
```

```python
l_y_pred = l_regressor.predict(X_test)
```

```python
l_r2 = r2_score(y_test, l_y_pred)
print(l_r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
r_regressor = RandomForestRegressor(n_estimators=10,random_state=0)
r_regressor.fit(X_train, y_train)
```

```python
r_y_pred = r_regressor.predict(X_test)
```

```python
r_r2 = r2_score(y_test, r_y_pred)
print(r_r2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_squared_error(y_test,r_y_pred)
print(mae)
```

```python
 mse = mean_squared_error(y_test,r_y_pred)
 print(mse)
```
