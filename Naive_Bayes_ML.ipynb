```python
# Exploring total_sqft feature
# The is_float function checks whether a given value can be converted
#to a float, returning True if it can and False if it cannot.
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None

# Apply the conversion function to the 'total_sqft' column
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)

# Drop rows with NaN values (if any) after conversion
df = df.dropna()
```

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
l_regressor = LinearRegression()
l_regressor.fit(X_train, y_train)
```

```python
l_y_pred = l_regressor.predict(X_test)
```

```python
l_r2 = r2_score(y_test, l_y_pred)
print(l_r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
r_regressor = RandomForestRegressor(n_estimators=10,random_state=0)
r_regressor.fit(X_train, y_train)
```

```python
r_y_pred = r_regressor.predict(X_test)
```

```python
r_r2 = r2_score(y_test, r_y_pred)
print(r_r2)
```

```python
import matplotlib.pyplot as plt # Import the matplotlib library

def plot_scatter_chart(df,total_sqft):
    bhk2 = df[(df.total_sqft==total_sqft) & (df.BHK==2)]
    bhk3 = df[(df.total_sqft==total_sqft) & (df.BHK==3)]
    plt.rcParams['figure.figsize'] = (8,6)  # Now use plt.rcParams
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(total_sqft)
    plt.legend()

plot_scatter_chart(df,2600)
```
